```python
import pandas as pd
```

## Читаем файл, записанный инференсером, который со скорами

```python
df = pd.read_csv('scores (1).csv')
```

```python
df['score'] = df['score'].apply(lambda x: int(x[2][0]))
```

```python
df['video_name'] = df['filename']
```

```python
def parser(df, delta=10, zero_delta=10):
    ans = []
    zeros = 0
    pred = 0
    now = 0
    start = True
    for i in range(len(df)):
        now = df.iloc[i]['frametime']
        if df.iloc[i].score == 1:
            zeros = 0
        if df.iloc[i].score == 0:
            zeros += 1
        if ((df.iloc[i].score == 0) and (zeros > zero_delta)) or ((now - pred > delta) and (df.iloc[i].score == 1)):
            start = True
        if ((df.iloc[i].score == 1) and start):
            ans.append(df.iloc[i].frametime)
            start = False
        pred = now
    return pd.Series(ans) #violation_timestamp под вставку, но его надо форматировать к timestamp
```

```python
df['violation_timestamp '] = parser(df)
```
